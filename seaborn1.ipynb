```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("tips")
```

##Corelation with heatmap

```python
 df = pd.get_dummies(df, columns=["sex", "day", "time", "smoker"])
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

Univarient/bivarient analysis
#jointplot

```python
 sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
 sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

Pairplot/scaterplot

```python
sns.pairplot(df)
```

```python
print(df.dtypes)
numerical_columns = df.select_dtypes(include=[np.number]).columns

# Create pair plot for numerical columns
sns.pairplot(df[numerical_columns])
```

```python
# Check data types and numerical columns
print(df.dtypes)
numerical_columns = df.select_dtypes(include=[np.number]).columns
# Create pair plot with 'sex' as hue
sns.pairplot(df, vars=numerical_columns, hue='sex_Male', diag_kind='kde')

# Display the plot
plt.show()
```
